```python
train_X = pd.read_csv("train_X.csv")
train_y = pd.read_csv("train_y.csv")
public_private = pd.read_csv("public_private_X.csv")
```

```python
all_data = pd.merge(train_X, train_y)
```

```python
train = all_data.drop(columns = ['ID', 'AVERAGE_DAILY_DEMAND_CASES', 'AVERAGE_VENDOR_ORDER_CYCLE_DAYS', 'AVERAGE_ORDER_CYCLE_CASES', 'PURCHASE_ORDER_DUE_DATE', 'ORDER_DATE', 'RESERVABLE_INDICATOR', 'TRANSIT_LEAD_TIME', 'DAYS_BETWEEN_ORDER_AND_DUE_DATE', 'PURCHASE_ORDER_TYPE'])
```

```python
train.columns
```

```python
X_train = train.drop(columns = 'ON_TIME_AND_COMPLETE')
y_train = train['ON_TIME_AND_COMPLETE']
```

```python
preprocessor = ColumnTransformer(
    transformers=[('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('poly', PolynomialFeatures(degree=2, include_bias=False)),
        ('scaler', StandardScaler())]), ['PURCHASING_LEAD_TIME', 'DISTANCE_IN_MILES', 'AVERAGE_ORDER_CYCLE_DAYS', 'ORDER_QUANTITY_DEVIATION', 'LEAD_TIME_TO_DISTANCE_RATIO', 'GIVEN_TIME_TO_LEAD_TIME_RATIO']),
        
        ('scale', StandardScaler(), ['AVERAGE_PRODUCT_ORDER_QUANTITY_MARKET', 'PRODUCT_MARKET']),
        
        ('cat', Pipeline([
            ('encoder', OneHotEncoder(handle_unknown='ignore'))
        ]), ['PRODUCT_CLASSIFICATION', 'DIVISION_CODE', 'DIVISION_NUMBER', 'DUE_DATE_WEEKDAY', 'ORDER_DAY_OF_WEEK', 'COMPANY_VENDOR_NUMBER', 'PURCHASE_FROM_VENDOR', 'PRODUCT_STATUS'])
    ])
```

```python
pipeline = Pipeline([
    ('preprocessor', preprocessor), 
    ('classifier', LogisticRegression(solver = 'liblinear', max_iter = 1000)) 
])

pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_train)
train_accuracy = accuracy_score(y_train, y_pred)
print(train_accuracy)
```

```python
import csv 

public_private['ON_TIME_AND_COMPLETE'] = pipeline.predict(public_private)

sk_submission = public_private[['ID', 'ON_TIME_AND_COMPLETE']]

sk_submission.to_csv('sk_submission.csv', index=False)
```
